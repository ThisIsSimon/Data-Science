# LASSO (NO PageValues)

```python
col = [col for col in df.columns if col not in ('PageValues', 'Revenue')]
x = df[col].values
scaler = StandardScaler()
scaler.fit(x)
x = scaler.transform(x)

y = df['Revenue'].values

list_of_coef = []

# MCCV
for i in range(1000):
    X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.8)
    clf = LogisticRegression(solver='liblinear', penalty='l1')
    clf.fit(X_train, y_train)


    list_of_coef.append([clf.coef_, clf.score(X_test, y_test)])
```

```python
lasso_df_norev = pd.DataFrame(list_of_coef, columns=['coef', 'accuracy'])

lasso_df_norev['accuracy'].mean()
```

```python
col = [col for col in df.columns if col not in ('PageValues', 'Revenue')]
average_coef = lasso_df_norev['coef'].mean()[0]
sns.barplot(y=average_coef, x=df[col].columns)
plt.title('LASSO Coefficients (No PageValues)')
plt.xticks(rotation=90)
plt.show()
```

## LASSO

```python
x = df[[col for col in df.columns if col != 'Revenue']].values
scaler = StandardScaler()
scaler.fit(x)
x = scaler.transform(x)

y = df['Revenue'].values

list_of_coef = []

# MCCV
for i in range(100):
    X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.8)
    clf = LogisticRegression(solver='liblinear', penalty='l1')
    clf.fit(X_train, y_train)


    list_of_coef.append([clf.coef_, clf.score(X_test, y_test)])
```

```python
lasso_df = pd.DataFrame(list_of_coef, columns=['coef', 'accuracy'])

lasso_df['accuracy'].mean()
```

```python
col = [col for col in df.columns if col not in ('Revenue')]
average_coef = lasso_df['coef'].mean()[0]
sns.barplot(y=average_coef, x=df[col].columns)
plt.title('LASSO Coefficients')
plt.xticks(rotation=90)
plt.show()
```
